```python
import numpy as np
import matplotlib.pyplot as plt

# Set the grid size that was used to generate the file
N = 20

# Load the eigenvalues from the file
# (Make sure the file "eigs_N20.txt" is in the same directory as this notebook)
eigenvalues = np.loadtxt(f"eigs_N{N}.txt")

# Plot the eigenvalues
plt.figure(figsize=(6, 4))
plt.plot(eigenvalues, 'o-', label='Eigenvalues')
plt.xlabel('Eigenvalue Index')
plt.ylabel('Energy (arbitrary units)')
plt.title(f'Eigenvalues for N = {N}')
plt.legend()
plt.show()
```

```python
# Define grid parameters for the potential plot
N_pot = 50  # Adjust as needed
x = np.linspace(-1, 1, N_pot)
y = np.linspace(-1, 1, N_pot)
X, Y = np.meshgrid(x, y)

# Define the harmonic potential: V(x,y) = 4*(x^2 + y^2)
V = 4. * (X**2 + Y**2)

plt.figure(figsize=(6, 5))
plt.imshow(V.T, origin='lower', extent=(-1, 1, -1, 1))
plt.colorbar(label='Potential Energy')
plt.title('Harmonic Potential V(x,y)')
plt.show()
```
